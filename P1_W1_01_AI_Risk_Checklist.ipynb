# Phase 1, Week 1: AI Risk Checklist

## Project

AI Risk Checklist for Datacenter Predictive Maintenance

## Objective

Create a practical AI risk register for an enterprise AI use case.

## Scenario

A datacenter wants to use an AI model to predict equipment failures, cooling problems, power anomalies, and maintenance requirements.

## Governance Goal

Identify key AI risks, score them, prioritize them, and define mitigation actions.

```python
import os
from datetime import date
import pandas as pd
import matplotlib.pyplot as plt

# Create output folder
os.makedirs("outputs", exist_ok=True)

# Risk register data
risks = [
    {
        "Risk ID": "R1",
        "Risk Type": "Data Quality",
        "Risk Description": "Sensor data may be incomplete, outdated, noisy, or inconsistent.",
        "Likelihood (1-5)": 4,
        "Impact (1-5)": 5,
        "Mitigation": "Perform data validation, missing value checks, and sensor calibration review.",
        "Owner": "Data Engineering Lead"
    },
    {
        "Risk ID": "R2",
        "Risk Type": "Model Accuracy",
        "Risk Description": "The model may incorrectly predict equipment failure or miss a critical event.",
        "Likelihood (1-5)": 4,
        "Impact (1-5)": 5,
        "Mitigation": "Use model performance monitoring, retraining, and human approval for high-risk alerts.",
        "Owner": "ML Engineer"
    },
    {
        "Risk ID": "R3",
        "Risk Type": "Operational Risk",
        "Risk Description": "AI recommendations may trigger unnecessary maintenance or delay urgent maintenance.",
        "Likelihood (1-5)": 3,
        "Impact (1-5)": 5,
        "Mitigation": "Keep human-in-the-loop approval for operational actions.",
        "Owner": "Operations Manager"
    },
    {
        "Risk ID": "R4",
        "Risk Type": "Cybersecurity",
        "Risk Description": "Attackers may manipulate sensor inputs or model outputs.",
        "Likelihood (1-5)": 3,
        "Impact (1-5)": 5,
        "Mitigation": "Apply access controls, anomaly detection, logging, and security monitoring.",
        "Owner": "Security Lead"
    },
    {
        "Risk ID": "R5",
        "Risk Type": "Compliance",
        "Risk Description": "Operational or personal data may be used without proper governance.",
        "Likelihood (1-5)": 2,
        "Impact (1-5)": 5,
        "Mitigation": "Classify data, anonymize sensitive data, and perform compliance review.",
        "Owner": "Compliance Officer"
    },
    {
        "Risk ID": "R6",
        "Risk Type": "Explainability",
        "Risk Description": "Teams may not understand why the AI predicted a failure.",
        "Likelihood (1-5)": 4,
        "Impact (1-5)": 4,
        "Mitigation": "Use explainable AI techniques and generate reason codes for predictions.",
        "Owner": "AI Governance Lead"
    },
    {
        "Risk ID": "R7",
        "Risk Type": "Vendor Risk",
        "Risk Description": "Third-party AI tools may create dependency, cost, or security concerns.",
        "Likelihood (1-5)": 3,
        "Impact (1-5)": 4,
        "Mitigation": "Perform vendor due diligence, SLA review, and exit strategy planning.",
        "Owner": "Vendor Manager"
    },
    {
        "Risk ID": "R8",
        "Risk Type": "Monitoring Failure",
        "Risk Description": "The AI model may degrade over time without detection.",
        "Likelihood (1-5)": 4,
        "Impact (1-5)": 4,
        "Mitigation": "Set up drift monitoring, alerting, and periodic model review.",
        "Owner": "MLOps Engineer"
    },
    {
        "Risk ID": "R9",
        "Risk Type": "Business Continuity",
        "Risk Description": "AI failure may impact datacenter uptime or maintenance planning.",
        "Likelihood (1-5)": 3,
        "Impact (1-5)": 5,
        "Mitigation": "Maintain fallback procedures and manual override processes.",
        "Owner": "Business Continuity Manager"
    },
    {
        "Risk ID": "R10",
        "Risk Type": "Governance Ownership",
        "Risk Description": "No clear accountability for AI decisions, incidents, or approvals.",
        "Likelihood (1-5)": 4,
        "Impact (1-5)": 4,
        "Mitigation": "Define RACI matrix, approval workflow, and AI governance board ownership.",
        "Owner": "AI Governance Architect"
    }
]

# Convert to DataFrame
df = pd.DataFrame(risks)

# Calculate risk score
df["Risk Score"] = df["Likelihood (1-5)"] * df["Impact (1-5)"]

# Define risk level
def risk_level(score):
    if score >= 20:
        return "Critical"
    elif score >= 12:
        return "High"
    elif score >= 6:
        return "Medium"
    else:
        return "Low"

df["Risk Level"] = df["Risk Score"].apply(risk_level)

# Sort by score
df = df.sort_values(by="Risk Score", ascending=False)

# Display result
display(df)

# Save CSV file
csv_file = "ai_risk_register_datacenter_pm.csv"
df.to_csv(csv_file, index=False)

print(f"CSV file created: {csv_file}")
```

```python
plt.figure(figsize=(12, 6))

plt.barh(
    df["Risk ID"] + " - " + df["Risk Type"],
    df["Risk Score"]
)

plt.gca().invert_yaxis()
plt.title("AI Risk Register: Datacenter Predictive Maintenance")
plt.xlabel("Risk Score = Likelihood x Impact")
plt.ylabel("Risk")
plt.tight_layout()

chart_file = "outputs/risk_score_chart.png"
plt.savefig(chart_file, dpi=150)

plt.show()

print(f"Chart saved: {chart_file}")
```
